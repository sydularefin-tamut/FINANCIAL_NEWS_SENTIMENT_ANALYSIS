```python
import pandas as pd
import numpy as np
import yfinance as yf
import seaborn as sns
import matplotlib.pyplot as plt
import spacy
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
from tqdm import tqdm

tqdm.pandas()

# Load News Dataset
news_file = r"C:\Users\SHAHBAZ\Desktop\reuters_headlines.csv"
news_df = pd.read_csv(news_file)
news_df.dropna(subset=['Description'], inplace=True)  # Remove rows with empty descriptions

# Initialize VADER sentiment analyzer
analyzer = SentimentIntensityAnalyzer()

# Perform sentiment analysis on headlines
def get_vader_sentiment(text):
    score = analyzer.polarity_scores(text)
    if score['compound'] >= 0.05:
        return 'POSITIVE'
    elif score['compound'] <= -0.05:
        return 'NEGATIVE'
    else:
        return 'NEUTRAL'

news_df["Sentiment"] = news_df["Headlines"].progress_apply(get_vader_sentiment)

# Visualize sentiment distribution
plt.figure(figsize=(8, 5))
sns.countplot(data=news_df, x="Sentiment", palette="coolwarm")
plt.title("Sentiment Distribution in Reuters Headlines")
plt.show()

# Load NLP model for named entity recognition (NER)
nlp = spacy.load("en_core_web_sm")

def extract_company(text):
    doc = nlp(text)
    for ent in doc.ents:
        if ent.label_ == "ORG":  # Organization names are likely companies
            return ent.text
    return None

# Process only top 100 rows
news_subset = news_df.head(100).copy()
news_subset["Extracted_Company"] = news_subset["Description"].apply(extract_company)
news_subset.dropna(subset=["Extracted_Company"], inplace=True)

def get_stock_ticker(company_name):
    try:
        ticker = yf.Ticker(company_name).history(period="1d")
        if not ticker.empty:
            return company_name  # If it returns data, assume it's a valid ticker
    except:
        return None
    return None

# Retrieve valid stock tickers
news_subset["Stock_Ticker"] = news_subset["Extracted_Company"].apply(get_stock_ticker)
news_subset.dropna(subset=["Stock_Ticker"], inplace=True)

def get_next_day_close(ticker, date):
    try:
        next_day = pd.to_datetime(date) + pd.Timedelta(days=1)
        stock_data = yf.Ticker(ticker).history(start=next_day.strftime('%Y-%m-%d'), end=(next_day + pd.Timedelta(days=1)).strftime('%Y-%m-%d'))
        if not stock_data.empty:
            return stock_data["Close"].values[0]
    except:
        return None
    return None

# Retrieve next day close price
news_subset["Next_Day_Close"] = news_subset.apply(lambda row: get_next_day_close(row["Stock_Ticker"], row["Time"]), axis=1)
news_subset.dropna(subset=["Next_Day_Close"], inplace=True)

# Analyze impact of sentiment on stock price
plt.figure(figsize=(10, 6))
sns.boxplot(data=news_subset, x="Sentiment", y="Next_Day_Close", palette="coolwarm")
plt.title("Impact of Sentiment on Next Day Close Price")
plt.show()

# Save results
news_subset.to_csv("news_sentiment_stock_analysis.csv", index=False)
print("Analysis completed and results saved!")

```
